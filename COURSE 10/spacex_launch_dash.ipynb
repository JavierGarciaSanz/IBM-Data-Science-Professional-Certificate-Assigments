```python
# --- SpaceX Launch Dashboard Static Visualization ---

import pandas as pd
import plotly.express as px

# Cargar el dataset (asegúrate de que el archivo está subido a Colab)
spacex_df = pd.read_csv("/content/drive/MyDrive/Data Science/spacex_launch_dash.csv")
spacex_df.head()
```

**Tasks 1 y 2**

```python
# Pie chart de éxitos totales por Launch Site
fig1 = px.pie(
    spacex_df,
    names="Launch Site",
    values="class",
    title="Total Successful Launches by Site"
)
fig1.show()

# Pie chart de éxito vs fallo para un sitio específico
site = "CCAFS LC-40"
site_df = spacex_df[spacex_df["Launch Site"] == site]

fig2 = px.pie(
    site_df,
    names="class",
    title=f"Success vs Failure for site {site}"
)
fig2.show()
```

**Tasks 3 y 4**

```python
# Rango de Payloads
min_payload = spacex_df["Payload Mass (kg)"].min()
max_payload = spacex_df["Payload Mass (kg)"].max()

# Filtrar por todo el rango de payloads
filtered_df = spacex_df[
    (spacex_df["Payload Mass (kg)"] >= min_payload)
    & (spacex_df["Payload Mass (kg)"] <= max_payload)
]

# Scatter para todos los Launch Sites
fig3 = px.scatter(
    filtered_df,
    x="Payload Mass (kg)",
    y="class",
    color="Booster Version Category",
    title="Correlation between Payload and Success for All Sites"
)
fig3.show()

# Scatter para un sitio específico
site = "KSC LC-39A"
site_df = filtered_df[filtered_df["Launch Site"] == site]

fig4 = px.scatter(
    site_df,
    x="Payload Mass (kg)",
    y="class",
    color="Booster Version Category",
    title=f"Correlation between Payload and Success for Site {site}"
)
fig4.show()
```
